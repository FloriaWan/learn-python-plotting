```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

# Heatmaps <br>
3-d data: histogram on x & y

```python
plt.figure()

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
_ = plt.hist2d(X, Y, bins=25)
```

```python
plt.figure()
_ = plt.hist2d(X, Y, bins=100)
```
